```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "D:/Matches/matches.csv"
ipl_df = pd.read_csv(file_path)

# Select relevant columns
ipl_data = ipl_df[['season', 'team1', 'team2', 'winner', 'city', 'player_of_match', 'win_by_runs']]

# Display the first few rows
print(ipl_data.head())
```

```python
# Basic Data Exploration

print(ipl_df.head())  # View the first few rows
print(ipl_df.info())  # Get information about columns and data types
```

```python
# Team Performance

team_wins = ipl_df['winner'].value_counts()
print("\nTeam Wins:\n", team_wins)
```

```python
# Player of the Match Analysis

player_of_match_counts = ipl_df['player_of_match'].value_counts()
print("\nPlayer of the Match Counts:\n", player_of_match_counts)
```

```python
# Runs Scored Analysis
#As Dataset Doesnt have runs column but having win_by_runs column so using that column

average_win_by_runs = ipl_df['win_by_runs'].mean()
print("\nAverage Win Margin (Runs):", average_win_by_runs)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Chart of Team Wins
plt.figure(figsize=(10, 6))
sns.countplot(x='winner', data=ipl_df) 
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')  
plt.xticks(rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram of Win Margins (Runs)
plt.figure(figsize=(8, 5))
plt.hist(ipl_df['win_by_runs'], bins=8, edgecolor='black')  
plt.title('Distribution of Win Margins (Runs)')
plt.xlabel('Win Margin (Runs)')
plt.ylabel('Frequency')  

plt.show()
```
